### 문제
N이 주어졌을 때, 1부터 N까지의 수로 이루어진 순열을 사전순으로 출력하는 프로그램을 작성하시오.

### 입력
첫째 줄에 N(1 ≤ N ≤ 8)이 주어진다. 

### 출력
첫째 줄부터 N!개의 줄에 걸쳐서 모든 순열을 사전순으로 출력한다.

```python
N = int(input())
arr = [i + 1 for i in range(N)]

def next_permutation(N, arr):
    idx = -1
    larr = len(arr)
    for i in range(larr -1, 0, -1):
        if arr[i] > arr[i-1]:
            idx = i-1
            break
    if idx == -1:
        return -1
    
    for j in range(larr-1, 0, -1):
        if arr[j] > arr[idx]:
            arr[idx], arr[j] = arr[j], arr[idx]
            arr[idx+1:] = arr[idx+1:][::-1] 
            return arr

print(*arr)
while True:
    arr = next_permutation(N, arr)
    if arr == -1:
        break
    print(*arr)
```
